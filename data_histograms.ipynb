```python
from coffea.util import load
from coffea import hist
import numpy as np
import matplotlib.pyplot as plt
import mplhep
```

```python
def create_plot1d(hist1d, lumi, log=False, ax=None, typ=None):
    
    from matplotlib.offsetbox import AnchoredOffsetbox, TextArea
    plt.style.use(mplhep.style.CMS)
    plt.rcParams.update({
        'font.size': 16,
        'axes.titlesize': 18,
        'axes.labelsize': 18,
        'xtick.labelsize': 14,
        'ytick.labelsize': 14
    })
    
    if typ == 'data':
        
        data_err_opts = {
        'linestyle': 'none',
        'marker': '.',
        'markersize': 10.,
        'color': 'k',
        'elinewidth': 1,
        }
        ax = hist.plot1d(hist1d, error_opts=data_err_opts)
        
        # CMS style
    
        CMS = plt.text(0.03, 1., "CMS",
                      fontsize=22,
                      fontweight='bold',
                      horizontalalignment='left',
                      verticalalignment='bottom',
                      transform=ax.transAxes
                     )
        pre = plt.text(0.13, 1.0, "Preliminary",
                      fontsize=19,
                      fontstyle='italic',
                      horizontalalignment='left',
                      verticalalignment='bottom',
                      transform=ax.transAxes
                     )
        lumi = plt.text(1., 1., r"" + lumi +  " fb$^{-1}$ (13 TeV)",
                    fontsize=18,
                    fontstyle='normal',
                    horizontalalignment='right',
                    verticalalignment='bottom',
                    transform=ax.transAxes
                   )
        
        
    elif typ == 'mc':
        
        fill_opts = {
        'alpha': 0.8,
        'edgecolor':(0,0,0,.5)
        }
        
        ax = hist.plot1d(hist1d, fill_opts=fill_opts)
        
        # CMS format
        hfont = {'fontname':'Helvetica'}    
        plt.text(0.13, 0.89, "CMS", fontdict=hfont,  fontweight='bold', transform=plt.gcf().transFigure)
        plt.text(0.19, 0.89, "Simulation", fontdict=hfont, style='italic', transform=plt.gcf().transFigure)
        
    else:
        
        data_err_opts = {
        'linestyle': 'none',
        'marker': '.',
        'markersize': 10.,
        'color': 'k',
        'elinewidth': 1,}
        
            
        ax = hist.plot1d(hist1d, ax=ax)#, error_opts=data_err_opts)

   
    ax.legend().remove()
    
    return ax
```

```python
import uproot
import hist
from hist import Hist
import matplotlib.pyplot as plt
import mplhep
```

```python
mplhep.style.use("CMS")
```

```python
#https://hist.readthedocs.io/en/latest/user-guide/notebooks/Plots.html
ttree = 'asso;1'
var_x = 'jpsi_dstar_deltarap'
name_x = 'rap'
label_x = r'$\Delta y_{J/\psi D*}$'
```

```python
def get_data(path_file, ttree, var_x,):
    
    with uproot.open(path_file) as f:
        #print(f.classnames())
        # Takes the ttree objec (defined in save_ttree.py script)
        tree_obj = f[ttree]
        # Takes its arrays
        branch_obj = tree_obj.arrays()
        
        # Takes the composed object inv mass
        # Takes the composed object delta y    
        jpsi_dstar_hist = branch_obj[var_x]
        
        return jpsi_dstar_hist
    
```

```python
bin_ht = 100
lumi='41.48' #2017
year='2017'

a = get_data('fit/data_root_files/2017/Charmonium2017_HLT_Dimuon25_dy_vtx0p05_sigma_eff.root', 'asso;1', 'jpsi_dstar_deltarap')
hist_1d_a = hist.Hist(hist.axis.Regular(bin_ht, 0, 5, name=name_x, label=label_x))   
hist_1d_a.fill(rap=a, weight=1)

b = get_data('fit/data_root_files/2017/Charmonium2017_HLT_Dimuon25_dstar_d0dca_cut.root', 'asso;1', 'jpsi_dstar_deltarap')
hist_1d_b = hist.Hist(hist.axis.Regular(bin_ht, 0, 5, name=name_x, label=label_x))   
hist_1d_b.fill(rap=b, weight=1.31)

fig, ax = plt.subplots(figsize=(18, 15))
mplhep.histplot(hist_1d_a, histtype='errorbar', label = 'Usual cuts')
mplhep.histplot(hist_1d_b, histtype='errorbar', label = 'DCA cut')

plt.legend()

ax.set_ylabel('Events/bin')

mplhep.cms.label('Preliminary', data=True, lumi=lumi, year=year[0:4])

#plt.savefig('jpsi_dstar_delta_rap.png')

plt.show()
```
